#### dataset 불러오기

```python
obj1 = "/content/drive/MyDrive/텍스트분석2023/traindata/objective_cv_train.txt"
obj2 = "/content/drive/MyDrive/텍스트분석2023/traindata/objective_test.txt"
sub1 = "/content/drive/MyDrive/텍스트분석2023/traindata/subjective_cv_train.txt"
sub2 = "/content/drive/MyDrive/텍스트분석2023/traindata/subjective_test.txt"

obj = []
sub = []
with open(obj1, "r") as f:
  example = f.readlines() # ['첫 번째 줄\n', '두 번째 줄\n', '세 번째 줄'] 저장
  for line in example:
    tmp = line.strip()
    obj.append(tmp.strip("."))

with open(obj2, "r") as f:
  example = f.readlines() # ['첫 번째 줄\n', '두 번째 줄\n', '세 번째 줄'] 저장
  for line in example:
    tmp = line.strip()
    obj.append(tmp.strip("."))

with open(sub1, "r") as f:
  example = f.readlines() # ['첫 번째 줄\n', '두 번째 줄\n', '세 번째 줄'] 저장
  for line in example:
    tmp = line.strip()
    sub.append(tmp.strip("."))

with open(sub2, "r") as f:
  example = f.readlines() # ['첫 번째 줄\n', '두 번째 줄\n', '세 번째 줄'] 저장
  for line in example:
    tmp = line.strip()
    sub.append(tmp.strip("."))


print(len(obj))
print(len(sub))

sub_label = [0]*len(sub)
obj_label = [1]*len(obj)

data = obj + sub
label = obj_label + sub_label
```

```python
import csv

col = ['message', 'label']

rows = zip(data, label)

with open('dataset.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(col)
    writer.writerows(rows)
```
